```python
#Import the modules
import requests
from bs4 import BeautifulSoup
```

```python
link="https://www.allrecipes.com/recipes/" #Define a link
page=requests.get(link) #Get the source file of the website
```

```python
page.content   #take a look at the content
```

```python
new=BeautifulSoup(page.content) #using BeautifulSoup to prettify the content and make it convenient for scraping
```

```python
new #take a look at it after prettfying it
```

```python
#try to get the string inside a single tag
new.find("h3",attrs={"class":"card__title"}).string.strip()  
```

```python
# save all the recipe title to alltitle
alltitle = new.find_all("h3",attrs={"class":"card__title"})
```
